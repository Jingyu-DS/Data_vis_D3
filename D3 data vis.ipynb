People are very sensitive to the price of goods so we chose to show how the price of wine is changing and how it is distributed. First, in this graph, we sort the prices of wine from highest to lowest. Admittedly, in linear scale, we can easily calculate the different prices and if we enlarge or reduce the the graph, the linear scale also enlarge or reduce in the same ratio. However, the graph looks imbalance because there are some outlier, which results that the most of point stay in one line. It's hard for us to tell the difference between them and get useful information in this graph. What's more, linear scale requires time and proficiency in drawing a graphic scale.

<img src="Log Scaling.png" width="700" height="300">

Log scale may be a good idea to overcome the disadvantages of linear scale. The big disadvantage of the log scale is that the data is negative. However, the price of wine is positive so we don't need to worry about it. The big difference between linear scale and log scale is that linear graphs are scaled so that equal vertical distances represent the same absolute dollar value change, the log scale shows percentage changes. For this reason, these point doesn't focus on the line. Obviously, on a log scale, we can easily know the distribution of the price of different kinds of wine, which will help us get the key information from the graph.

### Final Graph 
During our coding process, we remove the null price from the dataset, which might make the data size a little below the 200. We would like to focus our graph on the price of wine. There would be a transition of our graph between linear scale and log scale. The y-axis would be the price and the x-axis would be the index ordered by price. We use the average price of wine from different countries to color the points in the graph. The deeper the color of point is, the more expensive overall the wine in that country is. For linear scale, most of the points with the price below 100 dollars are at the bottom of the graph. Because of the log scale, lots of the points are scaled to the middle of the graph.

### Final Write-up
We try to make the graph more interactive. In our graph, the radio buttons are created to help the readers interact with our graph. They can click on either of the buttons to choose which scale they would like have a look. Also, the tooltips are included in the graph. When readers mouse over one of the points in the graph, the country and the price would be indicated there. The use of color is to indicate the average price of wine in different country.
```python
from IPython.display import HTML, Javascript, display

def configure_d3():
    display(Javascript("""
    require.config({
      paths: {
        d3: "https://d3js.org/d3.v6.min"
      }
    })"""))


configure_d3()
```

### Price of Wine

```python
%%html
<script type="text/javascript">   
require(['d3'], function (d3) {
    
    d3.csv('wine3.csv')
        .then(function(data) {
            d3.select("div#final").selectAll("*").remove()
            
            const width = 700
            const height = 400
            const margin = 60 
            data = data.filter(d=> (d.price != ""))
            data = data.map(d=> ({price:+d.price,location:d.country}))
            rollup = d3.rollup(data, v => d3.mean(v, d => d.price), k => k.location)
            var unroll = []
            for (let [key, value] of rollup) {
                unroll.push({"location":key,"avg":value})
            } 

            data = data.filter(d=> d.price != 0 )
            // Sorting by vaccinated and using a linear scale
            data = data.sort((a, b) => b.price - a.price)
            const xScale = d3.scaleLinear().range([margin , width - margin]).domain(d3.extent(data, (d,i) => i))
            const yScale = d3.scaleLinear().range([height-margin , margin]).domain(d3.extent(data, (d,i) => d.price))
            const color = d3.scaleLinear().range([0,1]).domain(d3.extent(unroll, (d,i) => d.avg))
            const palette = d3.interpolateReds
            
            const svg = d3.select("div#final").append("svg")
                .attr("width", width)
                .attr("height", height)
            
            
            const xAxis = d3.axisBottom().scale(xScale)
            
            svg.append("g")
                .attr("class", "axis")
                .attr("transform", "translate(0," + (height-margin) + ")")
                .call(xAxis) 
            
            svg.append("text")
                .attr("x", width/2)
                .attr("y", height-5)
                .style("text-anchor", "middle")
                .text("Index Ordered by Price")
            
            const yAxis = d3.axisLeft().scale(yScale).tickFormat((d,i) => d)
            
            svg.append("g")
                .attr("class", "yAxis")
                .attr("transform", "translate(" + margin + ",0)")
                .call(yAxis)  
            
            svg.append("text")
                .attr("transform", "rotate(-90,15,"+(height/2)+")")
                .attr("x", 15)
                .attr("y", height/2)
                .style("text-anchor", "middle")
                .text("Price of Wine (dollars)") 
            
            svg.append("line")
                .attr("")
            
            svg.selectAll("circle")
                .data(data)
                .join("circle")
                .attr("cx", (d,i)=> xScale(i))
                .attr("cy", (d,i)=> yScale(d.price))
                .attr("r",0)
                .transition()
                .duration(1000)
                .delay(1000)
                .attr("r", 5)
                .style("stroke","black")
                .style("fill", (d,i)=> palette(color(rollup.get(d.location))))
            svg.selectAll("circle")
                .append("title").text(d=> "Country: "+d.location+" Price($): "+d.price)
             
            
        })
        .catch(function(error){
        
        })
    
})
</script>
```

### Legend: Price of Wine

```python
%%html
<script type="text/javascript">   
require(['d3'], function (d3) {
    
    d3.csv('wine3.csv')
        .then(function(data) {
            d3.select("div#legend").selectAll("*").remove()
            const width = 800
            const height = 100
            const margin = 60 
            data = data.filter(d=> (d.price != ""))
            data = data.map(d=> ({price:+d.price,location:d.country}))


            data = data.sort((a, b) => b.price - a.price)
            const palette = d3.interpolateReds
            const scaling = d3.scaleLinear().range([margin,width-margin]).domain(d3.extent(data, (d,i) => d.price))
            
            const svg = d3.select("div#legend").append("svg")
                .attr("width", width)
                .attr("height", height)
            
            
            const xAxis = d3.axisBottom().scale(scaling)
            
            svg.append("g")
                .attr("class", "axis")
                .attr("transform", "translate(0," + (height-margin) + ")")
                .call(xAxis) 

            svg.append("text")
                .attr("x", width/2)
                .attr("y", height-5)
                .style("text-anchor", "middle")
                .text("Price of Wine")
                 
            
            const num = 20
            const values = d3.range(1,num)
            
            const coloring = d3.scaleLinear().range([0,1]).domain(d3.extent(values))
            var defs = svg.append("defs")
            var linearGradient = defs.append("linearGradient")
                .attr("id", "linear-gradient") 

            linearGradient.selectAll("stop").data(values).join("stop")
                .attr("offset", d=> d/num)
                .attr("stop-color", d=>palette(coloring(d)) )
            svg.append("rect")
                .attr("class","legend")
                .attr("x", margin)
                .attr("y", (height-margin)-50)
                .attr("width", (width-margin)-(margin))
                .attr("height", 50)
                .style("fill", "url(#linear-gradient)")

            
        })
        .catch(function(error){
            console.log(error)
        })
    
})
</script>
```

### Update Code

```python
%%html
<script type="text/javascript">  
require(['d3'], function (d3) {

    d3.select("div#input").selectAll("*").remove()
    d3.select("div#input").append("input")
        .attr("id","linear")
        .attr("type","radio")
        .attr("name","switch")
        .attr("value","linear")
        .attr("checked","true")
        .on("click", (event, d) => { switchGraph(event.currentTarget.value); })
    d3.select("div#input").append("label")
        .attr("for","switch")
        .text("Linear Scaling")
    d3.select("div#input").append("br")
    d3.select("div#input").append("input")
        .attr("id","log")
        .attr("type","radio")
        .attr("name","switch")
        .attr("value","log")
        .on("click", (event, d) => { switchGraph(event.currentTarget.value); })
    d3.select("div#input").append("label")
        .attr("for","switch")
        .text("Log Scaling")



    function switchGraph(x){
        d3.csv('wine3.csv')
            .then(function(data) {
        const height = 400
        const margin = 60 
        const width = 700
        data = data.filter(d=> (d.price != ""))
        data = data.map(d=> ({price:+d.price,location:d.country}))

        rollup = d3.rollup(data, v => d3.mean(v, d => d.price), k => k.location)
        var unroll = []
        for (let [key, value] of rollup) {
            unroll.push({"location":key,"avg":value})
        } 
        // Sorting by vaccinated and using a linear scale
        data = data.sort((a, b) => b.price - a.price)
        const yLinear = d3.scaleLinear().range([height-margin , margin]).domain(d3.extent(data, (d,i) => d.price))
        const yLog = d3.scaleLog().range([height-margin , margin]).domain(d3.extent(data, (d,i) => d.price))
        const colorLinear = d3.scaleLinear().range([0,1]).domain(d3.extent(unroll, (d,i) => d.avg))
        const colorLog = d3.scaleLog().range([0,1]).domain(d3.extent(unroll, (d,i) => d.avg))
        const palette = d3.interpolateReds
        
        if (x=="linear") {

            d3.selectAll("circle")
                .transition("circle")
                .delay((d,i)=> i*20)
                .duration(2000)
                .attr("cy", (d,i)=> yLinear(d.price))
                .style("fill", (d,i)=> palette(colorLinear(rollup.get(d.location))))
            const yAxis = d3.axisLeft().scale(yLinear).tickFormat((d,i) => d)

            d3.select(".yAxis")
                .transition("axis")
                .duration(1000)
                .call(yAxis)  

            const scaling = d3.scaleLinear().range([margin,width-margin]).domain(d3.extent(data, (d,i) => d.price))
            const xAxis = d3.axisBottom().scale(scaling)
            d3.select("div#legend").select(".axis")
                .transition("axis")
                .duration(1000)
                .call(xAxis) 


        }
        else {
            d3.selectAll("circle")
                .transition("circle")
                .delay((d,i)=> i*20)
                .duration(2000)
                .style("fill", (d,i)=> palette(colorLog(rollup.get(d.location))))
                .attr("cy", (d,i)=> yLog(d.price))
            const yAxis = d3.axisLeft().scale(yLog).tickFormat((d,i) => d )

            d3.select(".yAxis")
                .transition("axis")
                .duration(1000)
                .call(yAxis) 

            const scaling = d3.scaleLog().range([margin,width-margin]).domain(d3.extent(data, (d,i) => d.price))
            const xAxis = d3.axisBottom().scale(scaling)
            d3.select("div#legend").select(".axis")
                .transition("axis-legend")
                .duration(1000)
                .call(xAxis) 

        }
                


    })
}
        
})
</script>
```
